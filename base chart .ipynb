```r
all <- read.csv2(file="result_month_create_size_fork_forksize_active_activesize_activefork_activeforksize_sort", header = TRUE, dec = ".", fill = TRUE)
all
```

```r
#plot(all$date,all$create_num)
#str(all)
#all$date <- as.factor(all$date)
#str(all$date)
#lines(all$date,all$create_num
barplot(height=all$create_num, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$create_num)),
       main="month VS number of projects",xlab="month",ylab="create_num",width=1,space=0.3,col="blue",xpd=TRUE)
barplot(height=all$fork_num, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$fork_num)),
       main="month VS number of forks",xlab="month",ylab="fork_num",width=1,space=0.3,col="blue",,xpd=TRUE)
barplot(height=all$active_num, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$active_num)),
       main="month VS number of active",xlab="month",ylab="active_num",width=1,space=0.3,col="blue",,xpd=TRUE)
barplot(height=all$activefork_num, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$activefork_num)),
       main="month VS number of fork and active",xlab="month",ylab="active_fork_num",width=1,space=0.3,col="blue",,xpd=TRUE)
```

```r
barplot(height=all$create_size, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$create_size)),
       main="month VS total projects' size",xlab="month",ylab="size of all projects",width=1,space=0.3,col="blue",,xpd=TRUE)
barplot(height=all$fork_size, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$fork_size)),
       main="month VS fork projects' size",xlab="month",ylab="size of fork projects",width=1,space=0.3,col="blue",,xpd=TRUE)
barplot(height=all$active_size, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$active_size)),
       main="month VS active projects' size",xlab="month",ylab="size of active projects",width=1,space=0.3,col="blue",,xpd=TRUE)
barplot(height=all$activefork_size, names.arg=all$date,xlim=c(1,10 + nrow(all)),ylim=c(0,max(all$activefork_size)),
       main="month VS active and fork projects' size",xlab="month",ylab="size of active and fork projects",width=1,space=0.3,col="blue",,xpd=TRUE)
```
